```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps and min_samples as needed
y_dbscan = dbscan.fit_predict(X_scaled)
```

```python
supervised = dataset.copy()
supervised['Cluster_group'] = y_dbscan
```

```python
supervised.to_csv("cluster_spectral.csv", index=False)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], palette='deep')
plt.title('Clusters Identified by DBSCAN')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
