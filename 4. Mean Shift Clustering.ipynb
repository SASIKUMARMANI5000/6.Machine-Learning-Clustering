```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import MeanShift
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
mean_shift = MeanShift()
y_mean_shift = mean_shift.fit_predict(X)
```

```python
dataset['Cluster_group'] = y_mean_shift
```

```python
dataset.to_csv("cluster_meanshift.csv", index=False)
```

```python
centroids = mean_shift.cluster_centers_
```

```python
print("Cluster Centers:\n", centroids)
print("Number of clusters found:", len(np.unique(y_mean_shift)))
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_mean_shift, cmap='viridis', s=50, alpha=0.6, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title('Mean Shift Clustering Results')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.grid()
plt.show()
```
